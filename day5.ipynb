## PART 1

```python
# Process input
pages_after = dict()
printouts = []
for line in inputdata:
    if '|' in line:
        parts = line.rstrip().split('|')
        try:
            pages_after[parts[0]].append(parts[1])
        except KeyError:
            pages_after[parts[0]] = [parts[1]]
    if ',' in line:
        printouts.append(line.rstrip().split(','))
```

```python
# Solution
def check_printout(rules_dict, printout):
    previous_pages = set()
    for page in printout:
        barred = set(rules_dict[page])
        collisions = previous_pages.intersection(barred)
        if len(collisions) > 0:
            return 0
        previous_pages.add(page)
    mid_index = int((len(printout)-1)/2)
    return int(printout[mid_index])
    
```

```python
cumulative_pages = 0
for printout in printouts:
    cumulative_pages += check_printout(pages_after, printout)

print(cumulative_pages)
```

## PART 2

```python
# Get the complementary dictionary
pages_before = dict()
for key in pages_after.keys():
    for val in pages_after[key]:
        try:
            pages_before[val].append(key)
        except KeyError:
            pages_before[val] = [key]
```

```python
pages_before
```

```python
def fix_printout(pages_after, pages_before, printout):
    good_printout = [printout[0]]
    set_printout = set(printout[1:])
    while len(set_printout) > 0:
        page = set_printout.pop()
        # The page is placed before
        for index, good_page in enumerate(good_printout):
            if page in pages_before[good_page]:
                good_printout.insert(index, page)
                break
        # The page is already placed
        if page in good_printout:
            continue
        # The page is placed after
        for index, good_page in reversed(list(enumerate(good_printout))):
            if page in pages_after[good_page]:
                good_printout.insert(index+1, page)
                break
    return good_printout
```

```python
cumulative_pages = 0
for printout in  printouts:
    if check_printout(pages_after, printout) == 0:
        good_printout = fix_printout(pages_after, pages_before, printout)
        new_central_page = check_printout(pages_after, good_printout)
        assess = (new_central_page != 0)
        cumulative_pages += new_central_page

print(cumulative_pages)
```
