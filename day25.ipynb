```python
data_to_use = inputdata

locks = [data_to_use[i:i+7] for i, line in enumerate(data_to_use) if line.startswith("#####") and i % 8 == 0]
keys = [data_to_use[i:i+7] for i, line in enumerate(data_to_use) if line.startswith(".....") and i % 8 == 0]

print(locks)
print(keys)
```

# PART 1

```python
# Solve part 1
def get_number_code(entity):
    symbol = entity[0][0]
    code = []
    for column in range(len(entity[0])):
        for row in range(len(entity)):
            if entity[row][column] != symbol:
                code.append(row)
                break
    return tuple(code)

locks_codes = {}
keys_codes = {}

for ilock, lock in enumerate(locks):
    code = get_number_code(lock)
    for position, value in enumerate(code):
        if position not in locks_codes.keys():
            locks_codes[position] = {}
        for free_space in range(value, len(lock)):
            if free_space not in locks_codes[position].keys():
                locks_codes[position][free_space] = set()
            locks_codes[position][free_space].add(ilock)

valid_combinations = 0
for ikey, key in enumerate(keys):
    code = get_number_code(key)
    try:
        for position, value in enumerate(code):
            if position == 0:
                lock_set = locks_codes[position][value]
            else:
                lock_set = lock_set & locks_codes[position][value]
        valid_combinations += len(lock_set)
    except KeyError:
        continue

print(valid_combinations)
```
